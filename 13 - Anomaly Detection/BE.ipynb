```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd

data = pd.read_csv("data/dataset.csv")
data['day'] = data['day_cycle_window'].astype(str).str[0]
data['cycle']= data['day_cycle_window'].astype(str).str[2]
data['window']= data['day_cycle_window'].astype(str).str[4:]
data
```

```python
data[data["day_cycle_window"] == "1_1_1"]
```

```python
def plot_param(cycle, param):
    plt.plot(data[data["day_cycle_window"] == cycle][param])

def query(cycle,param):
    return(np.array(data[data["day_cycle_window"] == cycle][param]))
```

```python

for param in range(1,5): 
    for cycle in range(1,6):
        
        plt.figure(figsize=(30,8))
        for day in range(1,5):
            Y=[]
            for window in range(200):
                Y= np.concatenate((Y,(query(str(day)+'_'+str(cycle)+'_'+str(window),'p'+str(param)))))
            plt.plot(Y,label = 'day '+ str(day)) 
        plt.title('param'+str(param)+' cycle'+str(cycle))
        plt.legend()
```

```python
def plot_day(param, day,df):
    Y = []
    for i in range(10):
        for j in range(200):
            Y = np.concatenate(
                (
                    Y,
                    np.array(
                        df[
                            df["day_cycle_window"]
                            == str(day) + "_" + str(i) + "_" + str(j)
                        ][param]
                    ),
                )
            )
            # print(str(day)+'_'+str(i)+'_'+str(j))
        # print(i)
    plt.plot([i for i in range(len(Y))], Y)
    # print(np.shape(Y))
"""

for j in range(1, 12):
    plt.figure(figsize=(30, 8))
    for i in range(1, 5):
        plot_day("p" + str(j), i)"""
```

```python
data_count = data.groupby(["day_cycle_window"]).agg(['count'])['p1']

errors = data_count[data_count['count'] !=100]
errors.index

clean_data = data 
for i in errors.index:
    clean_data = clean_data.drop(clean_data.index[clean_data['day_cycle_window'] == i ])
clean_data
```

```python
for j in range(1, 12):
    plt.figure(figsize=(30, 8))
    for i in range(1, 5):
        plot_day("p" + str(j), i,clean_data)
```
